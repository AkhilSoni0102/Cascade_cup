```python
train.drop(columns=['followers_avg_age', 'num_of_comments', 'weekends_trails_watched_per_day', 'weekdays_trails_watched_per_day','slot1_trails_watched_per_day','slot2_trails_watched_per_day','slot3_trails_watched_per_day','slot4_trails_watched_per_day'], inplace=True)
test.drop(columns=['followers_avg_age', 'num_of_comments', 'weekends_trails_watched_per_day', 'weekdays_trails_watched_per_day','slot1_trails_watched_per_day','slot2_trails_watched_per_day','slot3_trails_watched_per_day','slot4_trails_watched_per_day'], inplace=True)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
import xgboost as xgb
from sklearn.metrics import f1_score
from sklearn.model_selection import GridSearchCV
```

```python
X= train.drop(columns=['age_group'])
Y= train['age_group']
X_train, X_test, y_train, y_test= train_test_split(X,Y, test_size=0.3, shuffle=True)
```

```python
print("X_train :", X_train.shape)
print("y_train :", y_train.shape)
print("X_test :", X_test.shape)
print("y_test :", y_test.shape)
```

## SVM

```python
SVC1= SVC(kernel='linear')
SVC2= SVC(kernel='rbf')
SVC3= SVC(kernel='poly')
SVC4= SVC(kernel='sigmoid')
svc_list= [SVC1,SVC2,SVC3,SVC4]
l=[]
for model in svc_list:
    model.fit(X_train, y_train)
    y_predict= model.predict(X_test)
    acc= f1_score(y_test, y_predict, average='weighted')
    l.append(acc)
l
```

## KNeighbors Classifier

```python
KNN= KNeighborsClassifier()
KNN_cv= GridSearchCV(KNN, param_grid={'n_neighbors': range(4,10,2)})
KNN_cv.fit(X_train, y_train)
KNN_predict= KNN_cv.predict(X_test)
KNN_accuracy= f1_score(y_test, KNN_predict, average='weighted')
print("KNN_accuracy: ", KNN_accuracy)
```

## Decision Tree Classifier

```python
DTC= DecisionTreeClassifier()
param= {'max_depth': range(4,12,2),'min_samples_split': range(1000,10000,500),'max_features': ['auto', 'sqrt', 'log2']}
DTC_cv= GridSearchCV(DTC, param_grid= param)
DTC_cv.fit(X_train, y_train)
DTC_predict= DTC_cv.predict(X_test)
DTC_accuracy= f1_score(y_test, DTC_predict, average='weighted')
print("DTC_accuracy: ", DTC_accuracy)
```

## Random Forest

```python
RFC= RandomForestClassifier(random_state=42, n_estimators= 500, max_depth=10, min_samples_split=2000)
RFC.fit(X_train, y_train)
RFC_predict= RFC.predict(X_test)
RFC_accuracy= f1_score(y_test, RFC_predict, average='weighted')
print("RFC_accuracy: ", RFC_accuracy)
```
